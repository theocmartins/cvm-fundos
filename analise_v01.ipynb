```python
import pandas as pd
import matplotlib.pyplot as plt


# Contar a frequência de eventos por data
frequencia = df['data'].value_counts().sort_index()

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
frequencia.plot(kind='bar', color='skyblue')
plt.title('Frequência de eventos por data')
plt.xlabel('Data')
plt.ylabel('Número de eventos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Agrupar por categoria e somar os aportes
soma_aportes = df.groupby('DENOM_SOCIAL')['aportes'].sum()

# Ordenar em ordem decrescente
soma_aportes = soma_aportes.sort_values(ascending=False)

# Separar os 10 maiores e somar os demais como 'Outros'
top_10 = soma_aportes[:10]
outros = soma_aportes[10:].sum()

# Adicionar 'Outros' ao final
top_10['Outros'] = outros

# Criar gráfico de pizza
plt.figure(figsize=(8, 8))
top_10.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Top 10 categorias por aportes + Outros')
plt.ylabel('')  # Remove o rótulo do eixo y
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Agrupar por categoria e somar os resgates
soma_resgates = df.groupby('DENOM_SOCIAL')['resgates'].sum()

# Ordenar em ordem decrescente
soma_resgates = soma_resgates.sort_values(ascending=False)

# Separar os 10 maiores e somar os demais como 'Outros'
top_10 = soma_resgates[:10]
outros = soma_resgates[10:].sum()

# Adicionar 'Outros' ao final
top_10['Outros'] = outros

# Criar gráfico de pizza
plt.figure(figsize=(8, 8))
top_10.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Top 10 categorias por resgates + Outros')
plt.ylabel('')  # Remove o rótulo do eixo y
plt.tight_layout()
plt.show()
```
